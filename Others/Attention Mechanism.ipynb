```python
import numpy as np
np.random.seed(0) 

class Node:
    def __init__(self, name):
        # node name
        self.name = name
        # the vector stored at this node
        self.data= np.random.randn(10)
        # weights governing how this node interacts with other nodes 
        self.wkey= np.random. randn (10, 10)
        self.wquery= np.random. randn(10, 10) 
        self.wvalue= np.random. randn(10, 10)

    def key(self):
        # what do I have?
        return self.wkey @ self.data
    
    def query(self):
        #what am I looking for?
        return self.wquery @ self.data
    
    def value(self):
        #what do I publicly reveal/broadcast to others? 
        return self.wvalue @ self.data
    
nodes = [Node(name) for name in range(5)]
# make 10 edges
randi = lambda: np.random.randint(len(nodes)) 
edges = [[randi(), randi()] for _ in range(10)]
updates = []
for i,node in enumerate(nodes):
    # what is this node looking for?
    query = node.query()

    # find all edges that are input to this node
    to_nodes = []
    for (from_node, to_node) in edges:
        if from_node == i:
            to_nodes.append(nodes[to_node])

    if len(to_nodes) == 0:
        continue # ignore

    # gather their keys, i.e. what they hold
    to_node_keys = [to_node.key() for to_node in to_nodes]

    # calculate the compatibilities 
    scores = [node_key.dot(query) for node_key in to_node_keys] 

    #softmax them so they sum to 1 
    scores = np.exp(scores)
    scores = scores / np. sum(scores)

    # gather the appropriate values with a weighted sum 
    values = [to_node_value.value() for to_node_value in to_nodes]
    update = sum([s * v for s, v in zip(scores, values)]) 
    updates.append(update)

for n,u in zip (nodes, updates):
    n.data = n.data + u # residual connection
```
